```python
import pandas as pd
import numpy as np
import tensorflow as tf
```

```python
# сразу зафиксируем сид для лучшей воспроизводимости результата
np.random.seed(17)
tf.random.set_seed(17)
```

```python
df = pd.read_csv('prepared_posts_df.csv')
```

```python
# создадим матрицы признаков и ответов
X = df.drop(columns=['rating']).to_numpy()
Y = df['rating'].to_numpy()
```

```python
A = (X.shape[1],)
```

```python
model = tf.keras.models.Sequential([
   tf.keras.layers.Dense(100, activation='relu', input_shape=A),
   tf.keras.layers.Dense(100, activation='relu'),
   tf.keras.layers.Dense(100, activation='relu'),
   tf.keras.layers.Dense(1,
activation=tf.keras.activations.linear)
])
model.summary()
```

```python
model.load_weights('homework_init_weights.h5') 
```

```python
model.compile(
    # # указываем функцию потерь
    loss=tf.keras.losses.MSE,
            metrics=[tf.keras.metrics.MeanAbsoluteError()]
             )
```

```python
#  1
```

```python
history = model.fit(X, Y)
```
